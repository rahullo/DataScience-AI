```python
# Import the necessary libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load the image
image_path = 'path_to_your_image.jpg'
image = cv2.imread('../IMAGES/CLina.jpg', cv2.IMREAD_GRAYSCALE)
image.shape
```

```python

# Function to apply log transformation
def log_transform(image):
    # Convert the image to float32 to avoid overflow or underflow during the log transformation
    image_float = np.float32(image)
    
    # Apply log transformation
    c = 255 / np.log(1 + np.max(image_float))
    log_image = c * np.log(1 + image_float)
    
    # Convert the image back to uint8
    log_image = np.uint8(log_image)
    
    return log_image
```

```python
# Apply log transformation
log_image = log_transform(image)

# Display the original and log-transformed images
plt.figure(figsize=(12, 6))

plt.subplot(1,2, 1)
plt.title('Original Image')
plt.imshow(image, cmap='gray')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title('Log Transformed Image')
plt.imshow(log_image, cmap='gray')
plt.axis('off')

plt.show()
```
