```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
# Load a noisy image
image = cv2.imread('../IMAGES/girlface.bmp', 0)  # Load as grayscale
```

```python
def median_filter(image, kernel_size):
    # Get the dimensions of the image
    m, n = image.shape
    
    # Create a copy of the image to store the output
    output_image = np.zeros((m, n), dtype=np.uint8)
    
    # Calculate the margin for the kernel size
    margin = kernel_size // 2
    
    # Iterate over the image (excluding the margins)
    for i in range(margin, m - margin):
        for j in range(margin, n - margin):
            # Extract the neighborhood
            neighborhood = image[i - margin:i + margin + 1, j - margin:j + margin + 1]
            
            # Flatten the neighborhood array and sort it
            flattened_neighborhood = neighborhood.flatten()
            sorted_neighborhood = np.sort(flattened_neighborhood)
            
            # Get the median value
            median_value = sorted_neighborhood[len(sorted_neighborhood) // 2]
            
            # Assign the median value to the output image
            output_image[i, j] = median_value
    
    return output_image

# Apply the custom median filter
kernel_size = 3
median_filtered_image = median_filter(image, kernel_size)

# Display original and filtered images
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1), plt.imshow(image, cmap='gray'), plt.title('Original Noisy Image')
plt.subplot(1, 2, 2), plt.imshow(median_filtered_image, cmap='gray'), plt.title('Median Filtered Image')
plt.show()
```

```python
# Apply median filter
median_filtered_image = cv2.medianBlur(image, 3)  # Kernel size of 5

# Display original and filtered images
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1), plt.imshow(image, cmap='gray'), plt.title('Original Noisy Image')
plt.subplot(1, 2, 2), plt.imshow(median_filtered_image, cmap='gray'), plt.title('Median Filtered Image')
plt.show()
```
