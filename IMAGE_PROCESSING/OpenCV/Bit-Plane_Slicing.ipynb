```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Read the image using OpenCV
image = cv2.imread('../IMAGES/zelda.bmp', 0)
image.shape
```

```python

def extract_bit_plane(image, bit):
    # Create a binary mask for the bit plane
    mask = 1 << bit
    # Extract the bit plane using bitwise AND and shift right operations
    bit_plane = (image & mask) >> bit
    # Scale the bit plane to the range [0, 255]
    bit_plane *= 255
    return bit_plane

# Extract and display all bit planes (0-7)
plt.figure(figsize=(12, 6))
for i in range(8):
    bit_plane = extract_bit_plane(image, i)
    plt.subplot(2, 4, i+1)
    plt.title(f'Bit Plane {i}')
    plt.imshow(bit_plane, cmap='gray')
    plt.axis('off')

plt.show()
```
