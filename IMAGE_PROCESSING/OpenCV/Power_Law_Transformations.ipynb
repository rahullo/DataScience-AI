```python
# Import necessary libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python

# Read the image using OpenCV
image = cv2.imread('../IMAGES/goldhill.bmp', cv2.IMREAD_GRAYSCALE)
image.shape
```

```python

# Check if the image was successfully loaded
if image is None:
    print("Error: Could not open or find the image.")
else:
    # Normalize the image to the range [0, 1]
    normalized_image = image / 255.0

    # Define the gamma value for the Power-law transformation
    gamma = 2.0  # You can change this value to see different results

    # Apply the Power-law transformation
    transformed_image = np.power(normalized_image, gamma)

    # Normalize the transformed image to the range [0, 255] and convert to uint8
    transformed_image = np.uint8(transformed_image * 255)

    # Display the original and transformed images using matplotlib
    plt.figure(figsize=(12, 6))

    # Original image
    plt.subplot(1, 2, 1)
    plt.title('Original Image')
    plt.imshow(image, cmap='gray')
    plt.axis('off')

    # Transformed image
    plt.subplot(1, 2, 2)
    plt.title(f'Power-law Transformed Image (Gamma = {gamma})')
    plt.imshow(transformed_image, cmap='gray')
    plt.axis('off')

    plt.show()
```
