```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
#image loading
image = cv2.imread('../IMAGES/man.bmp', 0)
image.shape
```

```python
#1. Write a program to perform image smoothening using an averaging filter of size 3 * 3, 15*15 and 51*51. Show results in a subplot.
def image_smoothening(image, kernel_size):
    # Hieght and width of the image
    height, width = image.shape

    # padding
    pad_size = kernel_size // 2
    padded_image = np.pad(image, pad_size, mode='constant', constant_values=0)

    # Initialize the output image
    smooth_image = np.zeros_like(image)

    # Iteration
    for i in range(height):
        for j in range(width):
            # matrix
            region = padded_image[i:i+kernel_size, j:j+kernel_size]
            
            # Compute the average value of the matrix
            average_value = np.sum(region) / (kernel_size * kernel_size)
            
            # Assign the average value to the corresponding pixel in the output image
            smooth_image[i, j] = average_value

    return smooth_image
```

```python
# Apply average smothening
kernel_size = 15  # Adjust kernel size as needed
smooth_image_3 = image_smoothening(image, 3)
smooth_image_15 = image_smoothening(image, 15)
smooth_image_51 = image_smoothening(image, 51)

plt.figure(figsize=(12, 4))

plt.subplot(1, 4, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 4, 2)
plt.imshow(smooth_image_3, cmap='gray')
plt.title('Smooth Image 1')
plt.axis('off')

plt.subplot(1, 4, 3)
plt.imshow(smooth_image_15, cmap='gray')
plt.title('Smooth Image 2')
plt.axis('off')

plt.subplot(1, 4, 4)
plt.imshow(smooth_image_51, cmap='gray')
plt.title('Smooth Image 3')
plt.axis('off')


plt.tight_layout()
plt.show()
```

```python
#3 Improve the image quality of the attached (Contrast.png image) using intensity level transformations of your choice. 
def contrast_stretching(image):
    # Minimum and maximum value o
    min_val = np.min(image)
    max_val = np.max(image)
    
    # Apply contrast stretching
    stretched_image = (image - min_val) * (255 / (max_val - min_val))
    stretched_image = np.clip(stretched_image, 0, 255).astype(np.uint8)
    
    return stretched_image
    
```

```python
# Apply contrast stretching
stretched_image = contrast_stretching(image)

#Plotting
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.xticks([]), plt.yticks([])

plt.subplot(1, 2, 2)
plt.imshow(stretched_image, cmap='gray')
plt.title('Contrast Stretched Image')
plt.xticks([]), plt.yticks([])

plt.show()
```
