```python
# Import necessary libraries
import cv2
import matplotlib.pyplot as plt
```

```python
# Load an image using OpenCV
image = cv2.imread('../IMAGES/cameraman.bmp', cv2.IMREAD_COLOR)
image.shape
```

```python
# Function to convert an image to its negative
def negative_image(image):
    # Maximum pixel value for 8-bit images is 255
    max_pixel_value = 255
    # Subtract image from the maximum pixel value to get the negative
    negative = max_pixel_value - image
    return negative
```

```python
# Check if the image is loaded successfully
if image is None:
    print("Error: Could not load image")
else:
    # Convert the image to its negative
    negative = negative_image(image)

    # Convert images from BGR (OpenCV format) to RGB (Matplotlib format)
    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    negative_rgb = cv2.cvtColor(negative, cv2.COLOR_BGR2RGB)

    # Display the original and negative images using Matplotlib
    plt.figure(figsize=(10, 5))

    plt.subplot(1, 2, 1)
    plt.title('Original Image')
    plt.imshow(image_rgb)
    plt.axis('off')

    plt.subplot(1, 2, 2)
    plt.title('Negative Image')
    plt.imshow(negative_rgb)
    plt.axis('off')

    plt.show()
```
