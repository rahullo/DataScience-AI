```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Read the input image
image = cv2.imread('../IMAGES/tank.bmp', cv2.IMREAD_GRAYSCALE)
```

```python
# Apply log transformation
c = 255 / np.log(1 + np.max(image))
log_transformed_image = c * (np.log(1 + image))
```

```python
# Convert the transformed image to uint8 type
log_transformed_image = np.array(log_transformed_image, dtype=np.uint8)
```

```python
# Display the original and log-transformed images
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(log_transformed_image, cmap='gray')
plt.title('Log Transformed Image')
plt.axis('off')

plt.tight_layout()
plt.show()
```
