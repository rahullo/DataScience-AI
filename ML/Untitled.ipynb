```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
health = pd.read_csv('DATA/2019.csv')
health.head()
```

```python
health['Country or region']
```

```python
health_2 = health[['Overall rank', 'Country or region', 'Score', 'GDP per capita']]
health_2.head()
```

```python
# Assuming 'x_column' is your X-axis and 'y_column' is your Y-axis
plt.plot(health_2['Score'], health_2['GDP per capita'])
plt.xlabel('X Axis Label')
plt.ylabel('Y Axis Label')
plt.title('Line Plot of X vs Y')
plt.show()
```

```python
# Simple bar plot
health_2.plot(kind='bar', x='Score', y='GDP per capita')
plt.xlabel('Score')
plt.ylabel('GDP per capita')
plt.title('Bar Plot of Categories')
plt.show()
```

```python
plt.scatter(health_2['Score'], health_2['GDP per capita'])
plt.xlabel('Life Satifaction')
plt.ylabel('GDP Per Capita')
plt.title('Scatter Plot of Life Satifaction vs GDP Per Capita')
plt.show()
```
